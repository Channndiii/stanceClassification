```python
import pandas as pd
import numpy as np
```

```python
def getFinalDf(task, Type):
    result_df = pd.read_csv('./data/result.csv')
    result_df = result_df[result_df.task == task]
    result_Type_df = result_df[result_df.type == Type]
    
    target_epoch = [str(i) for i in range(10, 101, 10)]
    result_Type_df['rowFilter'] = result_Type_df['fold_epoch'].apply(lambda fold_epoch: fold_epoch.split('_')[-1] in target_epoch)
    
    result_Type_df = result_Type_df[result_Type_df.rowFilter == True]
    
    result = np.asarray(result_Type_df.values[:, [4, 5, 6, 7, 8, 9]])
    final_df = pd.DataFrame(np.mean(result.reshape([5, -1, 6]), axis=0), 
                            columns=['train_acc', 'train_cost', 'test_acc', 'test_cost', 'disagree_f1', 'agree_f1'])
    def computeAverageF1(row):
        row['average_f1'] = (row['disagree_f1'] + row['agree_f1']) / 2
        return row
    final_df = final_df.apply(computeAverageF1, axis=1)
    return final_df
```
